```python
dataframes = {}
for folder in os.listdir(bronze_path):
    if folder == "desktop.ini":
        continue
    folder_path = os.path.join(bronze_path, folder)
    if os.path.isdir(folder_path):
        file = os.listdir(folder_path)[0]
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)
        dataframes[folder] = df
print(dataframes.keys())            
            
```

```python
silver_dataframes = {}

for table, df in dataframes.items():
    silver_dataframes[table] = df.copy()

print("Silver DataFrames Created Successfully!")
```

```python
date_columns = {
    "customers": ["contract_start_date"],
    "drivers": ["hire_date", "termination_date"],
    "delivery_events": ["event_timestamp"],
    "fuel_purchases": ["purchase_date"],
    "loads": ["load_date"],
    "maintenance_records": ["maintenance_date"],
    "routes": [],
    "safety_incidents": ["incident_date"],
    "trips": ["trip_date"],
    "trucks": ["purchase_date"],
    "truck_utilization_metrics": ["metric_date"],
    "driver_monthly_metrics": ["month"],
    "facilities": [],
    "trailers": ["purchase_date"]
}

for table, cols in date_columns.items():
    if table in silver_dataframes:
        for col in cols:
            if col in silver_dataframes[table].columns:
                silver_dataframes[table][col] = pd.to_datetime(
                    silver_dataframes[table][col],
                    errors="coerce"
                )

print("✅ Date columns converted successfully!")
```

```python
import os

silver_path = r"C:\Users\yashk\OneDrive\Desktop\Logistics_Data_Pipeline\Silver"

os.makedirs(silver_path, exist_ok=True)

for table, df in silver_dataframes.items():

    table_folder = os.path.join(silver_path, table)
    os.makedirs(table_folder, exist_ok=True)

    file_path = os.path.join(table_folder, f"{table}.csv")

    df.to_csv(file_path, index=False)

print("✅ All cleaned files saved successfully in separate folders!")
```
