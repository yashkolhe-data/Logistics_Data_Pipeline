```python
import pandas as pd
import os
```

```python
silver_path = r"C:\Users\yashk\OneDrive\Desktop\Logistics_Data_Pipeline\Silver"
```

```python
gold_dataframes = {}
for folder in os.listdir(silver_path):
    folder_path = os.path.join(silver_path, folder)
    if os.path.isdir(folder_path):
        file = os.listdir(folder_path)[0]
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)
        gold_dataframes[folder] = df
print(gold_dataframes.keys())
```

```python
drivers = gold_dataframes["drivers"]
driver_metrics = gold_dataframes["driver_monthly_metrics"]
drivers.head()
```

```python
print(drivers.columns.tolist())
print(driver_metrics.columns.tolist())
```

```python
drivers.head()
```

```python
driver_metrics.head()
```

```python
print(drivers.columns.tolist())
```

```python
print(driver_metrics.columns.tolist())
```

```python
driver_performance = pd.merge(
    drivers,
    driver_metrics,
    on="driver_id",
    how="left"
)
driver_performance.head()
```

```python
driver_performance.info()
```

```python
driver_summary = driver_performance.groupby(
    ["driver_id", "first_name", "last_name"]
).agg({
    "trips_completed": "sum",
    "total_miles": "sum",
    "total_revenue": "sum",
    "total_fuel_gallons": "sum",
    "average_mpg": "mean",
    "on_time_delivery_rate": "mean",
    "average_idle_hours": "mean"
}).reset_index()

driver_summary.head()
```

```python
gold_path = r"C:\Users\yashk\OneDrive\Desktop\Logistics_Data_Pipeline\Gold"
```

```python
import os

driver_summary_folder = os.path.join(gold_path, "driver_performance_summary")
os.makedirs(driver_summary_folder, exist_ok=True)

driver_summary.to_csv(
    os.path.join(driver_summary_folder, "driver_performance_summary.csv"),
    index=False
)

print("✅ Driver Performance Summary saved successfully!")
```

```python
routes = pd.read_csv("../Silver/routes/routes.csv")
loads = pd.read_csv("../Silver/loads/loads.csv")
```

```python
route_summary = loads.merge(routes, on="route_id", how="left")
```

```python
print(loads.columns.tolist())
```

```python
print(routes.columns.tolist())
```

```python
route_summary = loads.merge(routes, on="route_id", how="left")

route_summary = route_summary.groupby("route_id").agg({
    "load_id": "count",
    "revenue": "sum",
    "typical_distance_miles": "mean"
}).reset_index()

route_summary.rename(columns={
    "load_id": "total_loads",
    "revenue": "total_revenue",
    "typical_distance_miles": "average_distance"
}, inplace=True)

route_summary.head()
```

```python
import os

os.makedirs("../Gold/route_summary", exist_ok=True)

route_summary.to_csv(
    "../Gold/route_summary/route_summary.csv",
    index=False
)

print("✅ Route Summary saved successfully!")
```
